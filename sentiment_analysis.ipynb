```python
# Importo librerias necesarias
import pandas as pd
from textblob import TextBlob
```

```python
# Cargamos a un df el csv posterior al ETL
df = pd.read_csv('/content/drive/MyDrive/csv/user_reviews.csv',lineterminator='\n')
```

```python
# Creacion de la función sentiment_analysis
def sentiment_analysis(review):
    if isinstance(review, str):
        text = review  # El texto de la reseña se pasa directamente
        sentiment = TextBlob(text).sentiment.polarity

        if sentiment < -0.3:
            return 0  # Malo
        elif sentiment >= -0.3 and sentiment <= 0.3:
            return 1  # Neutral
        else:
            return 2  # Positivo
    else:
        return 1  # Valor predeterminado para reseñas no válidas o NaN
```

```python
# Aplicamos la función a la columna 'review' y se crea una nueva columna 'sentiment_analysis' que reempazara 'review'
df['sentiment_analysis'] = df['review'].apply(sentiment_analysis)
```

```python
df.drop(columns=['user_url'],inplace=True)
```
