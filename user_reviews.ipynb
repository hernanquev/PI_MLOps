```python
#Importamos librerias necesarias
import pandas as pd
import ast
import re
```

```python
# Inicializa una lista para almacenar los datos de las líneas del archivo JSON
rows = []

with open('/content/drive/MyDrive/australian_user_reviews.json', encoding = 'utf-8') as f:
  for line in f.readlines():
    rows.append(ast.literal_eval(line)) # Evalúa cada línea como un objeto JSON

# Crea un DataFrame a partir de la lista de datos
df_reviews = pd.DataFrame(rows)
```

```python
# Desanida la columna 'reviews' y crea un nuevo DataFrame
df_reviews = pd.json_normalize(rows, record_path='reviews', meta=['user_id','user_url'])
```

```python
df_reviews.head()
```

```python
df_reviews.drop(columns=['funny','last_edited','helpful'],inplace=True)
```

```python
df_reviews.drop_duplicates(subset=['user_id','item_id','review'],inplace=True)
```

```python
pattern = r"[^\w\s':]+"
df_reviews['review'] = df_reviews['review'].apply(lambda x: re.sub(pattern, '', x))
```

```python
df_reviews['posted'] = df_reviews['posted'].str.replace('Posted', '')
df_reviews['posted'] = df_reviews['posted'].str.replace('.', '')
df_reviews['posted'] = df_reviews['posted'].str.strip()
```

```python
# Guardamos los DataFrame en formato CSV
df_reviews.to_csv('/content/drive/MyDrive/csv/user_reviews.csv',index=False)
```
